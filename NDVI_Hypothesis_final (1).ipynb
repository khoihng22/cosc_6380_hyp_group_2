Vegetation growth is influenced by water availability; hence, areas with higher precipitation often exhibit denser vegetation. Satellite data, such as NDVI, can track this vegetation coverage, revealing a correlation with precipitation patterns. 


H0 (Null Hypothesis): There is no correlation between satellite vegetation coverage and monthly average precipitation. This suggests that changes in vegetation coverage are independent of variations in precipitation.​

H1 (Alternative Hypothesis): There is a significant correlation between satellite vegetation coverage and monthly average precipitation. This implies that vegetation coverage varies in response to changes in precipitation levels.​

## Data Extraction Process

For each NDVI file, the process involves selecting the nearest data point for each predefined coordinate and extracting the NDVI value. The extracted data includes the NDVI value, latitude, longitude, and the corresponding time. The results are compiled into a pandas DataFrame.


```python
import xarray as xr 
import pandas as pd
results = []

for file_path in file_paths:
    ds = xr.open_dataset(file_path)
    
    for idx, (lat, lon) in enumerate(coordinates):
        ds_point = ds.sel(lat=lat, lon=lon, method='nearest')
        ndvi_values = ds_point['NDVI'].values.flatten()
        times = ds_point['time'].values
        
        for time, ndvi in zip(times, ndvi_values):
            results.append({
                'location_index': idx,
                'NDVI_value': ndvi,
                'lat': lat,
                'long': lon,
                'time': pd.to_datetime(time).strftime('%Y-%m-%d')
            })

df_results = pd.DataFrame(results)
```

# Correlation Analysis between NDVI and Precipitation

This notebook explores the relationship between satellite-derived Normalized Difference Vegetation Index (NDVI) values and precipitation data. NDVI is a key indicator of vegetation health, while precipitation is a crucial environmental factor influencing vegetation patterns. Understanding the correlation between these variables can provide insights into how vegetation dynamics are influenced by rainfall patterns.


## Setup and Dependencies

First, let's import the necessary Python libraries for data manipulation, statistical analysis, and regression modeling. We'll use `pandas` for handling our dataset, `numpy` for numerical operations, `scipy.stats` for correlation analysis, and `statsmodels` for regression analysis.


```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
```

The dataset `ndvi_precip.csv` contains the NDVI and precipitation values. We'll load this data into a pandas DataFrame for analysis. Ensure that the file path matches the location of your dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='precipitation', y='NDVI', data=data, ci=None, scatter_kws={'alpha':0.5})
plt.title('NDVI vs. Precipitation')
plt.xlabel('Precipitation')
plt.ylabel('NDVI')
plt.grid(True)
plt.savefig('F:/PHD/Spring_2024/Data Analytics/Assignment_2/NDVI_analysis/correlation.png', dpi=300, bbox_inches='tight')
plt.show()
```

# Correlation Analysis

To understand the linear and monotonic relationships between NDVI and precipitation, we'll perform both Pearson and Spearman correlation analyses. These analyses will help us gauge the strength and direction of the association between these two variables.

The plot does not indicate a clear linear relationship between precipitation and NDVI. The points do not cluster around the regression line, which is relatively flat.

```python
# Pearson correlation
correlation = stats.pearsonr(data['precipitation'], data['NDVI'])
print(f'Pearson correlation coefficient: {correlation[0]}, p-value: {correlation[1]}')

# Spearman correlation
spearman_correlation = stats.spearmanr(data['precipitation'], data['NDVI'])
print(f'Spearman correlation coefficient: {spearman_correlation[0]}, p-value: {spearman_correlation[1]}')
```

Based on the provided correlation coefficients and p-values:

Pearson correlation coefficient: The value is very close to zero (0.000552) and the p-value is very high (0.991). This indicates a very weak linear relationship between NDVI and precipitation, and the high p-value suggests that this weak correlation is not statistically significant.

Spearman correlation coefficient: Similarly, the value is close to zero (-0.005) and the p-value is high (0.919). This indicates a very weak monotonic relationship between NDVI and precipitation, and again, the high p-value suggests this weak correlation is not statistically significant.

Interpretation:
Given the statistical evidence, we fail to reject the null hypothesis (H0). The data does not provide sufficient evidence to support a significant correlation between satellite vegetation coverage (NDVI) and monthly average precipitation. The Pearson and Spearman analyses suggest that changes in vegetation coverage, as measured by NDVI, are largely independent of variations in precipitation for the dataset analyzed.

This conclusion means, based on the data and analyses conducted, there is no strong evidence to suggest that areas with higher precipitation consistently exhibit denser vegetation as measured by NDVI, or vice versa, within the specific context and dataset analyzed. It's important to note that this conclusion is data-dependent and might vary with different datasets, regions, or analysis methods.

## Linear Regression Analysis

To further explore the relationship between NDVI and precipitation, we'll perform a linear regression analysis. This will allow us to quantify the change in NDVI with respect to changes in precipitation and assess the statistical significance of this relationship.


```python
X = sm.add_constant(data['precipitation'])  # Add a constant term for the intercept
model = sm.OLS(data['NDVI'], X).fit()  # Fit the model
print(model.summary())  # Print the summary statistics of the model
```
